```python
games_df = pd.read_csv("../Resources/vgsales.csv.zip")
list_of_games = games_df["Name"].values.tolist()
list_of_api_calls = []

browser = Browser("chrome", executable_path = "/usr/local/bin/chromedriver",headless=True)

for game in list_of_games[0:50]:
    browser.visit("https://www.youtube.com/")
    browser.fill("search_query", f"{game} trailer\n")
    browser.find_by_css("div#dismissable div.text-wrapper.style-scope.ytd-video-renderer div#meta div#title-wrapper h3.title-and-badge.style-scope.ytd-video-renderer").first.click()
    url = str(browser.url)
    title_search = re.findall(r'(?:https://www\.youtube\.com/watch\?v=)(.*)', url)
    videoId =title_search[0]
    YOUTUBE_API = f"https://www.googleapis.com/youtube/v3/commentThreads?part=snippet&maxResults=1000&videoId={videoId}&key=AIzaSyCh4fW4lDya8KjpB6g0eJ6UuS4gBmNErDc"
    results = requests.get(YOUTUBE_API).json()
    try:
        all_items = results["items"]
        list_of_comments = []
        game_dict = {}
        for comment in all_items:
            x= comment["snippet"]["topLevelComment"]["snippet"]["textOriginal"]
            list_of_comments.append(x)
        game_dict[game]= list_of_comments
        list_of_api_calls.append(game_dict)
        print("SUCCESS")
    except:
        print("FAILUREEEEEEEEEEEE")
        continue
    

```
