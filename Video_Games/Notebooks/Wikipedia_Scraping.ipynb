```python
practice_list = ["Call_of_Duty:_Advanced_Warfare"]
problem_movies = []
list_of_dicts = []

for name in practice_list:
    
    dictionary_of_info = {}
    
    browser.visit(f"https://www.wikipedia.org//wiki/{name.replace(' ', '_')}")
    html = browser.html
    my_soup = soup(html, "html.parser")
    try:
        game_table = my_soup.select_one("table.infobox tbody")
        my_rows = game_table.select("tr")

        for row in my_rows:
            title = row.select_one("th a")
            value = row.select_one("td div.plainlist ul li a")
            print(title)
            print(value)

            if title == None or value == None: 
                continue
            else:
                dictionary_of_info[title.get_text()]= value.get_text()
        list_of_dicts.append(dictionary_of_info)
    except:
        problem_movies.append(name)
```

```python
# Checkpoint
####
####

problem_movies = []
list_of_dicts = []

for name in list_of_games:
    
    dictionary_of_info = {}
    
    browser.visit(f"https://www.wikipedia.org//wiki/{name.replace(' ', '_')}")
    html = browser.html
    my_soup = soup(html, "html.parser")
    
    try:
        game_table = my_soup.select_one("table.infobox tbody")
        my_rows = game_table.select("tr")

        for row in my_rows:
            title_a = row.select_one("th a")
            title_th = row.select_one("th")
            check_value = row.select_one("td a")

            if (title_a == None and title_th == None) or check_value == None: 
                continue
            else:
                value = row.select_one("td")
                final_values = value.find_all("a")
                all_values = []
                for x in final_values:
                    all_values.append(x.get_text())
                if title_a != None: 
                    dictionary_of_info[title_a.get_text()]= all_values
                elif title_th != None:
                    dictionary_of_info[title_th.get_text()]= all_values
        list_of_dicts.append(dictionary_of_info)
        print("SUCCESSFUL")
    except:
        problem_movies.append(name)
        print("problem instance")
```
